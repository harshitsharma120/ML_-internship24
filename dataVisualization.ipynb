```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
sm=pd.read_csv('student_marks.csv')
```

```python
sm
```

```python
sm['Total']=sm[['English','Hindi','Maths','Science','SST']].sum(axis=1)
```

```python
sm
```

```python
sm['percentage']=sm[['Total']]/5
```

```python
sm
```

```python

A= 100-80
B= 80-60
C= 60-50
D= 50-40
E= 40-30
F=0-30
```

```python
def clac_grade(P):
    if P<=100 and P>=80:
        return 'A'
    elif P<80 and P>=60:
        return 'B'
    elif P<60 and P>=50:
        return 'C'
    elif P<50 and P>=40:
        return 'D'
    elif P<40 and P>=30:
        return 'E'
    elif P<30 and P>=0:
        return 'F'
    else:
        return 'X'
sm['Grade']=sm['percentage'].apply(clac_grade)
sm
```

```python
temp=sm.head()
sub=['English','Hindi','Maths','Science','SST']

plt.figure(figsize=(10,6))
plt.plot(temp['Name'],temp[sub],marker='o',label=sub)
plt.title("marks in defrent Subjects")
plt.xlabel('name')
plt.ylabel('marks')
plt.legend()
plt.grid(True)
plt.show()
```
